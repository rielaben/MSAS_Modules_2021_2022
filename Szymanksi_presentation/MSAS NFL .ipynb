```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
NFL=pd.read_excel("NFLW9_2021.xlsx")
NFL
```

```python
NFL['hptsdiff']= NFL['t1cumpfor']-NFL['t1cumpagn']
NFL['aptsdiff']=NFL['t2cumpfor']-NFL['t2cumpagn']
NFL
```

# Modeling results to week 7 using logistic regression

```python
NFL2to7 = NFL[(NFL['week']>1) & (NFL['week']<7)]
NFL2to7
```

```python
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
Win_GF = 'hwin~hptsdiff+aptsdiff'
model = smf.glm(formula = Win_GF, data = NFL2to7, family=sm.families.Binomial())
result = model.fit()
print(result.summary())
```

# Predicting week 8 results

```python
NFLW8 = NFL[NFL.week == 8]
NFLW8
```

```python
NFLW8['hwinprmod'] = 1/(1+np.exp(-(-.1972 +.0031*NFLW8['hptsdiff']-.0158*NFLW8['aptsdiff'])))
NFLW8
```

```python
NFLW8['hwinprbet'] = (1/NFLW8['t1odds'])/(1/NFLW8['t1odds']+1/NFLW8['t2odds'])
NFLW8
```

```python
NFLW8['Bet_true'] = np.where((NFLW8['hwinprbet']>.5) & (NFLW8['t1score']>NFLW8['t2score']),1, np.where((NFLW8['hwinprbet']<.5) & (NFLW8['t1score']<NFLW8['t2score']),1,0))
NFLW8
```

```python
NFLW8['Mod_true'] = np.where((NFLW8['hwinprmod']>.5) & (NFLW8['t1score']>NFLW8['t2score']),1, np.where((NFLW8['hwinprmod']<.5) & (NFLW8['t1score']<NFLW8['t2score']),1,0))
NFLW8
```
