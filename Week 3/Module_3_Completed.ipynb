1.   Which AFC East Quarterback had the most touchdowns per attempts?

```python
afc_east_list = ['NWE', 'NYJ', 'MIA', 'BUF']
afc_east_qbs = simple_df[(simple_df['Team'].isin(afc_east_list)) & (simple_df['Position']=='QB')]
afc_east_qbs['td_per_att'] = afc_east_qbs['Touchdowns']/afc_east_qbs['Att']
afc_east_qbs = afc_east_qbs.sort_values(by='td_per_att', ascending=False)
afc_east_qbs
```

2.   Filter the top 5 starting quarterbacks (started more han 12 games) based on Completion Percentage in 2 different age groups (<25, >33) and report which group has the higher average T5

*   filter out all players that didn't start >12 games
*   create two dataframes, one with quarterbacks age <25 and one with quarterbacks age >33
*   sort the two dataframes each by completion percentage
*   take the average of the completion percentage column of the first 5 rows and report the two averages

```python
starting_qbs = simple_df[(simple_df['GS']>12) & (simple_df['Position']=='QB')]

starting_qbs['cmp%'] = starting_qbs['Cmp']/starting_qbs['Att']

under_25 = starting_qbs[starting_qbs['Age'] < 25]
over_33 = starting_qbs[starting_qbs['Age'] > 33]

under_25_sorted = under_25.sort_values(by='cmp%', ascending=False).reset_index(drop=True)[:5]
over_33_sorted = over_33.sort_values(by='cmp%', ascending=False).reset_index(drop=True)[:5]

u25_average = under_25_sorted['cmp%'].mean()
over_33_average = over_33_sorted['cmp%'].mean()

display(under_25_sorted)
display(over_33_sorted)

print(f"under 25 average: {u25_average}")
print(f"over 33 average: {over_33_average}")
```
